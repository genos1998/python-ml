```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11) #number from 0 to 5 having 11 sample pt
print(x)
y=x**2
print(y)
```

```python
plt.figure()
```

```python
plt.figure(figsize=(30,25))
plt.plot(x,y,'.')
plt.plot(x,y,'*')
plt.plot(x,y)
plt.show()
```

```python
x= np.linspace(0,10,5)
y=x**2
plt.figure()
plt.plot(x,y,'*',color='#FF5733')
plt.plot(x,y,'-',color='g')
plt.xlabel('og numbers')
plt.ylabel('og numbers with values')
plt.title('first')
plt.grid()
#plt.axis([0,5,0,50])
plt.show()
```

```python
z=np.linspace(0,3,6)
y=z**2
plt.figure(figsize=[8,6],facecolor='y')
plt.plot(z,y,alpha=0.7,linestyle='-.',linewidth=3,marker='o',markersize=10)
plt.grid()
plt.xlabel('bhut')
plt.ylabel('yo')
plt.title('hi')
plt.show()
```

```python
x=np.linspace(0,4,8)
y=x+1
plt.subplot(2,3,1)
plt.plot(x,y,'r-.')
plt.subplot(2,3,2)
plt.plot(y,x,'c*')
plt.subplot(2,3,3)
plt.plot(y,x,'y.')
plt.subplot(2,3,4)
plt.plot(x,y,'g-.')
plt.subplot(2,3,5)
plt.plot(y,x,'b*')
plt.subplot(2,3,6)
plt.plot(y,x,'c.')
plt.show()
```

```python
x=np.linspace(2,10,16)
y=x+1
plt.figure()
y1=x+2
plt.plot(x,y,'r',label='add of 1')
plt.plot(x,y1,'k',label='add of 2')
plt.grid(color='k',alpha=0.5)
plt.legend(loc=10)
plt.show()
```

```python
x=[5,10,7]
y=[12,6,15]
x1=[6,11,10]
y1=[9,2,5]
plt.figure(facecolor='y')
plt.scatter(x,y,color='r',label='red',marker='*',s=200,alpha=0.7)
plt.scatter(x1,y1,color='k',label='black',marker='.',s=100,alpha=0.7)
plt.legend()
plt.grid()
plt.show()
```

```python
x=100+10*np.random.randn(10000)
print(x)
plt.hist(x,label='number of people in train',bins=100,cumulative=True) #bins means sample pt to be taken
plt.xlabel('people')
plt.ylabel('frequency in train')
plt.hist(x,label='number of people in train',bins=100,cumulative=False) 
#cumulative is false it won't plot cumulative graph
plt.legend()
plt.show()
```

```python
import numpy as jj
import matplotlib as plt1
x=jj.linspace(-50,150,30)
plt.subplot(3,3,1)
plt.plot(jj.sin(x))
plt.subplot(3,3,2)
plt.plot(jj.cos(x))
plt.subplot(3,3,3)
plt.plot(jj.sin(x)+jj.cos(x))
plt.subplot(3,3,4)
plt.plot(jj.cos(x))
plt.subplot(3,3,5)
plt.plot(jj.tan(x))
plt.subplot(3,3,6)
plt.plot(jj.tan(x)+jj.cos(x))
plt.subplot(3,3,7)
plt.plot(jj.tan(x))
plt.subplot(3,3,8)
plt.plot(jj.sin(x))
plt.subplot(3,3,9)
plt.plot(jj.tan(x)+jj.sin(x))
plt.show()
```

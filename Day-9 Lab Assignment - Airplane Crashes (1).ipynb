## 2 How many people were on board? Plot a line graph to describe the same and also find the Median value.

```python
import matplotlib.pyplot as plt
import numpy as np
df.head()
plt.figure(figsize=[20,10])
datanew=df[['Aboard','year']]
datanew.groupby('year')['Aboard'].sum().plot()
med=datanew.groupby('year')['Aboard'].sum().median()
print("The Median value is ",med)
plt.xticks(np.arange(1908,2010,1),rotation=90)
plt.yticks(np.arange(0,4000,100))
plt.xlabel("Year")
plt.ylabel("Number of people On Board")
plt.title("Number of peoples OnBoard  vs  Year")
plt.show()
```

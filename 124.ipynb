```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python

cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'flower']
dataset=pd.read_csv('iris.data.txt', header=None, names=cols)
dataset
```

```python
iris_setosa=dataset.loc[dataset["flower"]=="Iris-setosa"]
iris_virginica=dataset.loc[dataset["flower"]=="Iris-virginica"]
iris_versicolor=dataset.loc[dataset["flower"]=="Iris-versicolor"]
```

```python
iris_setosa
```

```python
sepal_length=iris_setosa[['sepal_length']]
sepal_length
```

```python
sepal_width=iris_setosa[['sepal_width']]
sepal_width
```

```python
from sklearn.model_selection import train_test_split
```

```python
sepal_length_train,sepal_length_test,sepal_width_train,sepal_width_test=train_test_split(sepal_length ,sepal_width,test_size=1/3,random_state=0)
```

```python
plt.scatter(sepal_length_train,sepal_width_train,color='red')

plt.title('sepal length vs sepal width')

```

```python
plt.scatter(sepal_length_test,sepal_width_test,color='blue')
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit([sepal_length_train],sepal_width_train)
```

```python
plt.plot(sepal_length_train,regressor.predict(sepal_length_train),color='blue')
```

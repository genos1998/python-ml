```python
from itertools import permutations
m=0
x=int(input())
s=int(input())
y=int(input())
m=int(input())
z=int(input())
c=int(input())
totalqt=x+y+z
t=''
v=''
j=''
u=''
l=''
k=''
if totalqt==3:
      strr='ABC'
elif totalqt==4:
      strr='ABCD'
elif totalqt==5:
      strr='ABCDE'
elif totalqt==6:
      strr='ABCDEF'
elif totalqt==7:
      strr='ABCDEFG'
elif totalqt==8:
      strr='ABCDEFGH'
elif totalqt==9:
      strr='ABCDEFGHI'
elif totalqt==10:
      strr='ABCDEFGHIJ'
elif totalqt==11:
      strr='ABCDEFGHIJK'
elif totalqt==12:
      strr='ABCDEFGHIJKL'
elif totalqt==13:
      strr='ABCDEFGHIJKLM'
elif totalqt==14:
      strr='ABCDEFGHIJKLMN'
elif totalqt==15:
      strr='ABCDEFGHIJKLMNO'
elif totalqt==16:
      strr='ABCDEFGHIJKLMNOP'
elif totalqt==17:
      strr='ABCDEFGHIJKLMNOPQ'
elif totalqt==18:
      strr='ABCDEFGHIJKLMNOPQR' 
elif totalqt==19:
      strr='ABCDEFGHIJKLMNOPQRS'
elif totalqt==20:
      strr='ABCDEFGHIJKLMNOPQRST'
elif totalqt==21:
      strr='ABCDEFGHIJKLMNOPQRSTU'
elif totalqt==22:
      strr='ABCDEFGHIJKLMNOPQRSTUV'
elif totalqt==23:
      strr='ABCDEFGHIJKLMNOPQRSTUVW'
elif totalqt==24:
      strr='ABCDEFGHIJKLMNOPQRSTUVWX'
elif totalqt==25:
      strr='ABCDEFGHIJKLMNOPQRSTUVWXY'
else:
      strr='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
for a in range(0,x):
    t=t+strr[a]
for b in range(x,x+y):
    v=v+strr[b]
for p in range(x+y,totalqt):
    j=j+strr[p]
totalletter=s+m+c
no=input()
no=no.replace(' ','')
one=input()
for f in range(0,s):
    u=u+t[f]
for w in range(0,m):
    k=k+v[w]
for r in range(0,c):
    l=l+j[r]
if one[0] and one[1] in u:
        output=x*y*z
        print(output)
        print(7)
elif one[0] and one[1] in k:
        output=x*y*z
        print(output)
        print(7)
elif one[0] and one[1] in l:
        output=x*y*z
        print(output)
        print(7)
else:
    output2=totalletter-len(one)-len(no)
    output=x*y*z
    print(output)
    print(output2)
```
